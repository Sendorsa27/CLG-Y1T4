## **NumPy**

**NumPy** (Numerical Python) is a fundamental package for numerical computation in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of high-level mathematical functions to operate on these arrays.

### **Key Concepts in NumPy**

Let's explore some core concepts that are frequently used when working with NumPy arrays.

```python
import numpy as np # Alias
```

#### **1. Data Loading (`np.loadtxt`)**

NumPy offers various functions to load data from text files. `np.loadtxt()` is a common one used to load data from a text file into a NumPy array. You can specify the file path and the data type (`dtype`) for the elements.

```python
# Create a dummy text file for demonstration
with open('dummy_data.txt', 'w') as f:
    f.write('1 2 3\n4 5 6\n7 8 9')

# Load data from the dummy file
example_data = np.loadtxt('dummy_data.txt', dtype='int')
# 1. I am loading a text file called dummy_data.txt
# 2. I am telling python that the data type of its content is integer.
print("Loaded data:\n", example_data)
```

#### **2. Array Attributes (`.shape`, `.ndim`)**

- **`.shape`**: This attribute returns a tuple indicating the size of the array in each dimension. For example, a `(96, 6)` shape means 96 rows and 6 columns.
- **`.ndim`**: This attribute returns the number of dimensions of the array (e.g., 1 for a 1D array, 2 for a 2D array).

```python
# Using the example_data from above
print("Shape of array:", example_data.shape)
print("Number of dimensions:", example_data.ndim)
```

#### **3. Indexing and Slicing**

NumPy arrays support powerful indexing and slicing techniques to access specific elements, rows, or columns:

- **Basic Indexing**: Accessing individual elements or rows (e.g., `array[0]`).
- **Slicing**: Extracting sub-arrays by specifying a range of indices (e.g., `array[:5]` to get the first 5 rows).
- **Advanced Indexing (Column Extraction)**: Using `array[:, column_index]` allows you to select an entire column from a 2D array. For instance, `array[:, 0]` fetches all elements from the first column.

```python
example_array = np.array([[10, 11, 12], [13, 14, 15], [16, 17, 18]])
print("Original array:\n", example_array)

# Basic Indexing
print("First row (basic indexing):", example_array[0])

# Slicing
print("First two rows (slicing):\n", example_array[:2])

# Advanced Indexing (Column Extraction)
print("First column:\n", example_array[:, 0])
```

#### **4. Transposing Arrays (`.T`)**

The `.T` attribute allows you to transpose an array, swapping its rows and columns. If an array has shape `(m, n)`, its transpose will have shape `(n, m)`.

```python
print("Original array shape:", example_array.shape)
print("Transposed array:\n", example_array.T) # np.transpose
print("Transposed array shape:", example_array.T.shape)
```

#### **5. Data Type Conversion (`np.array(..., dtype=...)`)**

When loading data, NumPy might infer a general data type (like string). You can explicitly convert the data type of an array or a specific column using `np.array(your_array, dtype='desired_type')` (e.g., `'int'`, `'float'`). This is crucial for performing numerical operations.

```python
string_numbers = np.array(['1', '2', '3', '4'])
print("Original string array:", string_numbers, "with dtype:", string_numbers.dtype)

# Convert to integer type
int_numbers = np.array(string_numbers, dtype='int')
print("Converted integer array:", int_numbers, "with dtype:", int_numbers.dtype)
```

#### **6. Aggregate Functions (`np.mean`, `np.sum`, `np.min`, `np.max`, `np.argmin`, `np.argmax`, `np.unique`)**

NumPy provides many functions for summarizing data:

- **`np.mean()`**: Calculates the arithmetic mean.
- **`np.sum()`**: Adds all elements in the array.
- **`np.min()` / `np.max()`**: Finds the minimum and maximum values.
- **`np.argmin()` / `np.argmax()`**: Returns the indices of the minimum and maximum values.
- **`np.unique()`**: Finds unique elements and optionally their counts.

```python
numeric_array = np.array([10, 20, 30, 40, 50])
print("Numeric array:", numeric_array)
print("Sum:", np.sum(numeric_array))
print("Mean:", np.mean(numeric_array))
print("Max value:", np.max(numeric_array))
print("Index of max value:", np.argmax(numeric_array))

categorical_array = np.array(['red', 'blue', 'red', 'green', 'blue', 'red'])
unique_elements, counts = np.unique(categorical_array, return_counts=True)
print("\nUnique elements:", unique_elements)
print("Counts:", counts)
```

#### **Question:** How would you find the minimum value in `numeric_array`? How would you find out if there are any duplicate elements in `categorical_array` without using `np.unique`?

#### **7. Sorting (`np.sort`, `np.argsort`)**

NumPy provides tools to sort array data efficiently:

- **`np.sort()`**: Returns a sorted copy of an array.
- **`np.argsort()`**: Returns the indices that would sort an array. This is extremely useful when you want to use the order of one array to sort another.

```python
unsorted_array = np.array([30, 10, 50, 20, 40])
print("Unsorted array:", unsorted_array)

# Basic Sorting
sorted_array = np.sort(unsorted_array)
print("Sorted array:", sorted_array)

# Using argsort
indices = np.argsort(unsorted_array)
print("Indices that would sort the array:", indices)
print("Using indices to sort manually:", unsorted_array[indices])
```

#### **8. Masking (Boolean Indexing)**

Masking, also known as Boolean indexing, is a powerful feature in NumPy. It allows you to select elements from an array based on a condition. You create a boolean array (mask) where `True` indicates elements that satisfy the condition, and then use this mask to index another array.

For example, `array[condition]` will return all elements from `array` where `condition` is `True`.

```python
data_to_filter = np.array([10, 20, 30, 40, 50, 60])
print("Original data:", data_to_filter)

# Create a boolean mask
condition = data_to_filter > 35
print("Condition (data > 35):", condition)

# Apply the mask
filtered_data = data_to_filter[condition]
print("Filtered data (elements > 35):", filtered_data)
```
