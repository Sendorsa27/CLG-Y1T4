It's evident that `director_id` in `movies` corresponds to `id` in `directors`. Therefore, we will use these two columns as our **keys** to merge the datasets.

### Understanding `isin()` for Membership Testing

The `.isin()` method is used to check if each element in a Series is contained in a provided list or Series of values. It's a very efficient way to perform membership testing across an entire column.

**How is `isin()` different from Python's standard `in` operator?**
- Python's `in` operator typically works for checking membership of **a single element** within a collection.
- Pandas' `.isin()` method performs this check for **all values** in a Series simultaneously, making it highly efficient for DataFrame operations.

As you can see, the output of `movies['director_id'].isin(directors['id'])` is a **boolean mask**. This is a Series of `True`/`False` values, where `True` indicates that the `director_id` at that position is found in the `directors['id']` column, and `False` if it's not.

**How can we easily check if *any* `False` values exist in this boolean mask?** (i.e., if there's any `director_id` in `movies` that *doesn't* have a match in `directors`)

````python
directors['id']
````

````python
movies['director_id'].isin(directors['id'])
````

````python
# Check if ALL director_id values inmovies['director_id'].isin(directors['id']) 'movies' are present in the 'id' column of 'directors'
# np.all() returns True if all elements in the boolean Series are True.
np.all(movies['director_id'].isin(directors['id']))
````

### Deciding on the Merge Strategy: Left Join

Now that we've confirmed the relationship between our two datasets, let's merge them. The choice of merge type is crucial and depends on what we want to achieve:

- **Do we need to keep all the rows from the `movies` DataFrame?** Yes, because we want to analyze all movies.
- **Do we need to keep all the rows from the `directors` DataFrame?** No. We only need the director information for those directors who actually directed a movie in our `movies` dataset.

Based on these requirements, a **`LEFT` Join** is the most appropriate merge type. A left join keeps all rows from the 'left' DataFrame (our `movies` DataFrame) and matches them with rows from the 'right' DataFrame (our `directors` DataFrame) based on the specified keys (`director_id` and `id`). If a match isn't found in the right DataFrame, the columns from the right DataFrame will have `NaN` (Not a Number) values for that row.

````python
# Perform a left merge of 'movies' and 'directors' DataFrames
# 'left_on' specifies the column from the left DataFrame ('movies')
# 'right_on' specifies the column from the right DataFrame ('directors')
# 'how='left'' ensures all rows from 'movies' are kept.
data = movies.merge(directors, how='left', left_on='director_id',right_on='id')
# Display the merged DataFrame
data
````

### Understanding and Cleaning Merged Columns (`id_x`, `id_y`)

Upon merging, you might have noticed two new columns: `id_x` and `id_y`.

**What are these newly created columns?**
When two DataFrames are merged and both contain columns with the same name (in this case, `id`), Pandas automatically renames them to prevent conflicts. By default, it appends `_x` to the column from the left DataFrame (`movies`) and `_y` to the column from the right DataFrame (`directors`).

- `id_x` represents the unique ID for each movie from the original `movies` DataFrame.
- `id_y` represents the `id` of the director from the `directors` DataFrame that matched the `director_id` in the `movies` DataFrame.

**Are any of these columns redundant and can be safely removed?**
- The `id_y` column essentially duplicates the `director_id` column after the merge, as it contains the same director ID used for the join. Once the merge is complete, `id_y` doesn't add new, unique information beyond what `director_id` already provides.
- While `director_id` served its purpose for the merge, we might not need it for subsequent analysis if we're now working with `director_name`.

Thus, we can safely drop both `director_id` (since we have `director_name` linked) and `id_y` to keep our DataFrame clean and avoid redundant information.

````python
# Drop the redundant 'director_id' and 'id_y' columns from the merged DataFrame
# 'axis=1' specifies that columns should be dropped
# 'inplace=True' modifies the DataFrame directly without needing re-assignment
data.drop(['director_id','id_y'],axis=1,inplace=True)
# Display the first few rows of the cleaned DataFrame
data.head()
````

````python
data.rename(columns={"id_x": "movies_id"}, inplace=True)
````

## Mastering the `.apply()` Method: Custom Transformations

The `.apply()` method is an incredibly versatile function in Pandas that allows you to apply a custom function along either axis (rows or columns) of a DataFrame or Series. This is particularly useful when you need to perform operations that aren't built-in Pandas functions.

**Scenario:** Let's say we want to convert the categorical values in the `gender` column (e.g., 'Male', 'Female') into a numerical format. This process is called **encoding**, and it's a common step in data preparation for machine learning models.

**Goal:**
- Encode 'Male' as `0`
- Encode 'Female' as `1`

**How can we encode the values in the `gender` column using a custom function?**
First, let's write a simple Python function that performs this encoding for a single value.

````python
data.head()
````

````python
# Define a function to encode gender values
def encode(data):
  if data == "Male":
    return 0  # Encode 'Male' as 0
  elif data == "Female":
    return 1  # Encode 'Female' as 1
  else:
    return "Not Found"
````

Now that we have our `encode` function, **how can we apply this function to *every* value in the `gender` column of our `data` DataFrame?** The `.apply()` method is perfect for this.

````python
# Apply the 'encode' function to the 'gender' column
# This transforms 'Male'/'Female' strings into 0/1 integers
data['gender'] = data['gender'].apply(encode)
# Display the updated DataFrame to see the encoded 'gender' column
data
````

````python
"""
Monday = 0
Tuesday = 1
Wednesday = 2
Thursday = 3
Friday = 4
Saturday = 5
Sunday = 6
"""

# 1. Create a function
def encode_days(data):
  if data == "Monday":
    return 0
  elif data == "Tuesday":
    return 1
  elif data == "Wednesday":
    return 2
  ...


data["day"].apply(encode_days)
````

### Applying Functions on Multiple Columns: Understanding `axis`

Notice how using `.apply()` with our custom `encode` function is conceptually similar to `Vectorization` in NumPy, where an operation is applied efficiently across an entire array or Series.

**Question:** What if we want to apply a function that involves *multiple* columns? For example, how can we find the sum of `revenue` and `budget` for each movie?

If we try to apply a sum directly to multiple columns, as shown in the previous cell, we might get an unexpected result. We want the sum *per movie* (i.e., per row), but the default behavior sums *down* the columns.

````python
# def profit()
````

````python
data[["revenue", "budget"]]
````

````python
# Attempt to apply the sum function to 'revenue' and 'budget' columns
# By default, apply() with axis=0 sums down each column separately
data[['revenue', 'budget']].apply(np.sum, axis=0)
````

### The Importance of the `axis` Parameter

As you saw, applying `np.sum` to `data[['revenue', 'budget']]` without specifying the `axis` parameter resulted in the sum of all `revenue` values and the sum of all `budget` values separately (summing along `axis=0`, which is the default for columns).

To achieve our goal of summing `revenue` and `budget` *for each movie* (i.e., *per row*), we need to explicitly tell Pandas to apply the function along `axis=1`.

**Understanding `axis` in `.apply()`:**
- **`axis=0`**: Applies the function **column-wise**. This is the default. The function receives each column as a Series.
- **`axis=1`**: Applies the function **row-wise**. The function receives each row as a Series.

By setting `axis=1`, our `np.sum` function now receives each row (containing `revenue` and `budget` for a single movie) and correctly calculates their sum.

**Your Turn:** Similarly, how can you calculate the `profit` for each movie (where `profit = revenue - budget`) and add it as a new column to our DataFrame? Think about how to define a function that takes a row and returns the profit.

````python
# Apply the sum function across rows (axis=1) for 'revenue' and 'budget'
# This calculates the sum of revenue and budget for each movie
data[['revenue', 'budget']].apply(np.sum, axis=1)
````

By setting the `axis=1`, every row of `revenue` was added to same row of `budget`.

**What does this `axis` mean in apply?**
- `axis=0` $\rightarrow$ It will apply to **each column**
- `axis=1` $\rightarrow$ It will apply to **each row**
  
Note that **by default, axis=0**.

**Similarly, how can I find the `profit` per movie (revenue-budget)?**

````python
x = data.loc[2]
````

````python
x['revenue'] - x['budget']
````

````python
# Define a function to calculate profit for a given row (x)
def prof(x):
  # Profit is calculated as revenue minus budget
  return x['revenue']-x['budget']

# Apply the 'prof' function to each row (axis=1) of the 'revenue' and 'budget' columns
# Store the result in a new 'profit' column
data['profit'] = data[['revenue', 'budget']].apply(prof, axis=1)
# Display the updated DataFrame including the new 'profit' column
data
````

## The Power of `groupby()`: Split, Apply, Combine

The `groupby()` method is one of the most powerful and frequently used functions in Pandas. It allows you to group rows together based on common values in one or more columns, and then perform operations on these groups independently.

The process of `groupby()` can be conceptualized as three distinct steps:

1.  **Split**: Pandas divides the DataFrame into smaller

## Group-based Operations: Aggregation

After splitting the DataFrame into groups using `groupby()`, the next step is often to **Apply** an aggregation function to each group. Aggregation functions condense the data in each group into a single summary statistic.

Common aggregate functions include:
-   `mean()`: Calculates the average of values in each group.
-   `sum()`: Calculates the total sum of values in each group.
-   `min()`: Finds the minimum value in each group.
-   `max()`: Finds the maximum value in each group.
-   `count()`: Counts the number of non-null items in each group.

**Let's begin by grouping our `data` DataFrame by `director_name`.** This will allow us to analyze metrics for each director.

### Group based Aggregation

We use different aggregate functions like `mean`, `sum`, `min`, `max`, `count` etc. on columns while grouping.

Let's group our data director-wise.

````python
data.head()
````

````python
data.groupby('director_name')
````

````python
# Group the DataFrame by 'director_name'
# This creates a DataFrameGroupBy object, not a DataFrame itself
data.groupby('director_name')["profit"].mean() # director -> 1 -> 1 Million USD ,  0 USD; 2 -> 500K, 500k, 500k, 500K
````

````python
data.groupby('director_name')["movies_id"].count().sort_values(ascending=False)
````

### Understanding the `DataFrameGroupBy` Object

When you call `data.groupby('director_name')`, it doesn't immediately return a new DataFrame. Instead, it returns a `DataFrameGroupBy` object. This object is an intermediate step that holds the grouping information but hasn't yet applied any operations.

- It's a special object that knows *how* to group your data.
- It's **not a DataFrame** itself; you need to apply an aggregation or transformation function to it to get a DataFrame or Series back.

**Question:** How can we find out the total number of distinct groups (i.e., the number of unique directors) our data has been divided into?

````python
# Get the number of distinct groups (unique directors) after grouping
data.groupby('director_name').ngroups # Number of Groups
````

### Inspecting Group Contents: `groups` and `get_group()`

After understanding the number of groups, we can also inspect the actual composition of these groups. The `.groups` attribute of a `DataFrameGroupBy` object returns a dictionary where keys are the group names (e.g., director names) and values are the indices of the rows belonging to that group.

**Question:** Based on this grouping, how can we see which specific rows (movies) belong to a particular director's group? For instance, how would we view all movies directed by 'Alexander Payne'?

````python
# View the 'groups' attribute of the groupby object
# This shows a dictionary where keys are group names and values are row indices
data.groupby('director_name').groups
````

**What if we want to extract data of a particular group from this list?**

````python
# Extract all rows (movies) belonging to the 'Alexander Payne' group
data.groupby('director_name').get_group('Steven Spielberg')
````

### Counting Movies per Director

Now, let's use aggregation to answer a practical question: **How can we find the total count of movies directed by each director?**

To do this, we can select a column (like `title`) after grouping by `director_name` and then apply the `.count()` aggregation function. This will count the non-null movie titles for each director.

````python
data.groupby('director_name')['profit'].mean()
````

````python
# Group by 'director_name' and count the number of 'title' entries for each director
# This effectively counts the number of movies directed by each director
data.groupby('director_name')['title'].count()
````

### Multiple Aggregations with `.aggregate()` (or `.agg()`)

Sometimes, you need to calculate several different summary statistics for the same grouped data. Pandas' `.aggregate()` (or its alias `.agg()`) method allows you to apply multiple aggregation functions at once.

**Scenario:** Let's find the very first year and the latest year a director released a movie. This involves finding the minimum (`min`) and maximum (`max`) values of the `year` column for each `director_name` group.

**Note:** Both `.aggregate()` and `.agg()` perform the same function. `.agg()` is often preferred for its conciseness.

````python
data.groupby("director_name").agg({"title" : "count", "profit" : "mean", "year": "min"})
````

````python
data.groupby("director_name")["profit"].aggregate(["min", "max", "sum", "mean"])
````

````python
# Group by 'director_name' and calculate the minimum and maximum 'year' for each director
# .aggregate() allows applying multiple functions to a column within groups
data.groupby(['director_name'])["year"].aggregate(['min', 'max'])
````

````python
data.groupby(['director_name'])["year"].aggregate(['min', 'max']).apply(lambda x : x["max"] - x["min"], axis=1)
````

````python
data.groupby("director_name")["popularity"].agg(["mean", "min", "max", "count"])
````

**Note:** We can also use `.agg` instead of `.aggregate` (both are same)

````python
# Group the data by 'director_name' and perform multiple aggregations
director_stats = data.groupby('director_name').agg(
    total_revenue=('revenue', 'sum'),
    average_vote=('vote_average', 'mean'),
    movie_count=('title', 'count')
)

# Sort the results by total_revenue in descending order
director_stats = director_stats.sort_values(by='total_revenue', ascending=False)

# Display the top directors based on these metrics
print("\n--- Director Performance Statistics (Sorted by Total Revenue) ---")
display(director_stats.head(10))
````

## Group-based Filtering: Isolating Specific Groups

Group-based filtering allows us to select entire groups of data based on a condition that applies to the group as a whole, rather than filtering individual rows. This is different from `boolean masking`, which filters individual rows.

**Scenario:** Our film studio client wants to identify "high-budget directors." Let's define a "high-budget director" as any director who has directed **at least one movie with a budget greater than or equal to 100 million USD**.

To achieve this, we can follow a multi-step process or use the powerful `.filter()` method:

1.  **Step 1: Aggregate Max Budget per Director**
    First, we need to find the maximum budget for each director. We can **group** the `data` by `director_name` and then use the `max()` aggregation on the `budget` column.

1. We can **group** the data by director and use `max` of the budget as aggregator.

````python
# Group data by 'director_name' and find the maximum 'budget' for each director
# .reset_index() converts the grouped result back into a DataFrame
data_dir_budget = data.groupby("director_name")["budget"].max().reset_index()
# Display the first few rows of the resulting DataFrame
data_dir_budget.head()
````

2.  **Step 2: Identify High-Budget Director Names**
    Next, from our `data_dir_budget` DataFrame, we need to **filter** out the names of directors whose maximum movie budget was greater than or equal to 100 million USD. We can use boolean indexing for this.

````python
# Filter 'data_dir_budget' to get 'director_name' for directors whose maximum movie budget was greater than or equal to 100 million USD
names = data_dir_budget.loc[data_dir_budget["budget"] >= 100000000, "director_name"]
````

3.  **Step 3: Retrieve Movies by High-Budget Directors**
    Finally, with the list of `names` (our high-budget directors), we can filter our original `data` DataFrame to retrieve all movies directed by these specific individuals using the `.isin()` method.

````python
# Filter the original 'data' DataFrame to include all movies directed by the 'high-budget' directors identified in the previous step
# The .isin(names) method checks if each director's name in 'data' is present in our list of 'names'
data.loc[data['director_name'].isin(names)]
````

### Simplified Group Filtering with `.filter()`

The previous three-step process to filter groups is effective, but Pandas offers a more concise way to achieve the same result using the `.filter()` method directly on the `groupby` object. This method is specifically designed for group-based filtering.

**Question:** Can we filter groups in a single, elegant step using a `lambda` function with `.filter()`?

Yes! The `.filter()` method takes a function (often a `lambda` function for conciseness) that operates on each group. If the function returns `True` for a group, that entire group is included in the result; otherwise, it's excluded.

Let's try to achieve the same "high-budget director" filtering using `groupby().filter()`.

````python
# Use groupby().filter() to select entire groups (directors) where the maximum budget of their movies is >= 100
# The lambda function defines the condition for filtering each group (x represents a group's DataFrame)
data.groupby('director_name').filter(lambda x: x["budget"].max() >= 100)
````

### Understanding `groupby().filter()`

Notice what's happening here:
-   We first create a `groupby` object based on `director_name`.
-   Then, we use the `.filter()` method, passing it a `lambda` function. This function is applied to *each individual group*.
-   The `lambda x: x["budget"].max() >= 100` condition checks if the maximum budget within that specific director's group (`x`) is at least 100 million USD.
-   If the condition evaluates to `True` for a group, that entire group (all rows associated with that director) is retained in the resulting DataFrame. If `False`, the entire group is dropped.

This is known as **Group-Based Filtering** because we are making a decision about whether to keep or discard an entire group based on a characteristic of that group.

**Key Points:**
-   The `.filter()` method operates on **groups**, not individual rows.
-   The result of `groupby().filter()` is **not a `groupby` object**, but a regular **Pandas DataFrame** containing only the groups that satisfied the filtering criterion.

## Group-based `apply()`: Performing Transformations within Groups

While `groupby()` with `.aggregate()` is great for summarizing data, and `groupby()` with `.filter()` is for selecting groups, `groupby()` with `.apply()` allows for more complex, arbitrary transformations on each group. This means you can run a custom function on each sub-DataFrame (group) independently and then combine the results.

**Scenario:** Our client wants to identify "risky movies." A movie is considered "risky" if its `budget` was higher than the *average `revenue` generated by that director's other movies*. This requires a calculation that uses an aggregate (average revenue) *within* each director's group, but then applies that to individual movies within the same group.

**How can we subtract the average `revenue` of a director from the `budget` column for each movie directed by that director, and then flag movies as 'risky'?**

We will define a custom function that takes a group (a sub-DataFrame for a single director) and calculates a new column based on group-specific aggregations.

````python
# Define a function to determine if a movie is 'risky' within a director's group
def func(x):
  # A movie is 'risky' if its budget is greater than or equal to the director's average revenue
  x["risky"] = x["budget"] - x["revenue"].mean() >= 0
  return x

# Group the data by 'director_name' and apply the 'func' function to each group
# 'group_keys=False' prevents the group key (director_name) from becoming part of the index
data_risky = data.groupby("director_name", group_keys=False).apply(func)
````

### Deconstructing `groupby().apply()`

Let's break down what we did with `groupby().apply()`:

1.  **Custom Function Definition (`func`):** We defined a function `func(x)` that takes a DataFrame `x` (which, when used with `groupby().apply()`, will be each sub-DataFrame corresponding to a director).
    - Inside `func`, we calculate the mean `revenue` for that specific director's group (`x["revenue"].mean()`).
    - Then, for *each movie* within that group, we compare its `budget` to the director's average `revenue`. If `budget - director_avg_revenue >= 0`, the movie is flagged as `True` (risky); otherwise, `False`.
    - The function then returns the modified group DataFrame `x`.

2.  **`groupby("director_name", group_keys=False).apply(func)`:**
    - We group the `data` DataFrame by `director_name`.
    - `group_keys=False`: This parameter prevents Pandas from adding the group key (`director_name` in this case) as an index level in the output. Setting it to `False` keeps the output DataFrame structure cleaner. (Note: The default behavior of `group_keys` might change in future Pandas versions).
    - The `.apply(func)` then takes our `func` and applies it to *each* `director_name` group. Pandas passes each sub-DataFrame `x` to our `func`.

**What was the overall process here?**
-   We defined a custom function (`func`) that encapsulates our specific business logic.
-   We grouped the `data` by `director_name`.
-   We used the `.apply()` method with our custom function on the grouped data, allowing us to perform a calculation that combines both individual row data (`budget`) with group-level aggregates (mean `revenue`).

Now, let's see if there are any movies that our analysis has flagged as "risky" based on this criterion.

````python
# Filter the 'data_risky' DataFrame to show only the movies flagged as 'risky' (where the 'risky' column is True)
data_risky.loc[data_risky["risky"]]
````
